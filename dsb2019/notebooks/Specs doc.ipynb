```python
def print_specs(specs):
    old = pd.options.display.max_colwidth 
    pd.options.display.max_colwidth = 999
    try:
        for i, event_id, info, args in specs.itertuples():
            display(Markdown(f"""__{event_id}__


{info}"""))
            display(pd.DataFrame(data=literal_eval(args)))
    finally:
        pd.options.display.max_colwidth = old


def find_common_parameters(specs):
    parameters = reduce(set.__and__, specs.args.apply(lambda args: set(pd.DataFrame(literal_eval(args)).name.values)).values)
    return parameters
        

def filter_specs(specs, test):
    filtered_specs = specs[specs.event_id.isin(test.event_id.unique())].copy()
    print(len(specs) - len(filtered_specs), "events never occur in test")
    return filtered_specs



print_specs(specs)
```
